```python
#First retrive States & it's corresponding Gross State Domestic Product(Renamed to GSDP).
state_total_df = all_states_df.loc[(all_states_df.Item == "Gross State Domestic Product")][['2014-15','State']]
state_total_gdp = state_total_df.rename(columns={'2014-15':'GSDP'})
state_total_gdp
```

```python
#State's Primary GSDP retriving:
state_primary = all_states_df.loc[(all_states_df.Item == "Primary")][['2014-15','State']]
state_primary = state_primary.rename(columns={'2014-15':'Primary_GSDP'})
state_primary
```

```python
#State's Secondary GSDP retriving:

state_Secondary = all_states_df.loc[(all_states_df.Item == "Secondary")][['2014-15','State']]
state_Secondary=state_Secondary.rename(columns={'2014-15':'Secondary_GSDP'})
state_Secondary
```

```python
#State's Tertiary GSDP retriving:
state_Tertiary = all_states_df.loc[(all_states_df.Item == "Tertiary")][['2014-15','State']]
state_Tertiary=state_Tertiary.rename(columns={'2014-15':'Tertiary_GSDP'})
state_Tertiary
```

```python
#Merging 'primary' & 'Secondary'
prim_sec_df = pd.merge(state_primary,state_Secondary, on = 'State')
#Merging 'primary_Secondary' & 'Tertiary'
prim_sec_ter_df = pd.merge(prim_sec_df,state_Tertiary,on = 'State')
prim_sec_ter_df
```

```python
#Merger of all the 2 df's: 'Primary_secondary_state_Tertiary' + Total:

total_gdp_all_sectors = pd.merge(prim_sec_ter_df,state_total_gdp, on = 'State')
total_gdp_all_sectors
```

```python
#Calculating the Primary,Secondary,Tertiary Percentage Contributions:
total_gdp_all_sectors['Pri_%_Contrib']= total_gdp_all_sectors['Primary_GSDP']/total_gdp_all_sectors['GSDP'] * 100
total_gdp_all_sectors['Sec_%_Contrib']= total_gdp_all_sectors['Secondary_GSDP']/total_gdp_all_sectors['GSDP'] * 100
total_gdp_all_sectors['Ter_%_Contrib']= total_gdp_all_sectors['Tertiary_GSDP']/total_gdp_all_sectors['GSDP'] * 100
total_gdp_all_sectors['tot_Pri_Sec_Ter_%_Contrib'] = total_gdp_all_sectors['Pri_%_Contrib'] + total_gdp_all_sectors['Sec_%_Contrib'] + total_gdp_all_sectors['Ter_%_Contrib']
total_gdp_all_sectors.sort_values(by='tot_Pri_Sec_Ter_%_Contrib',ascending=False)
total_gdp_all_sectors
```

```python
#Selecting the States,Primary,Secondary,Tertiary Percentage Contributions:
primary = total_gdp_all_sectors['Pri_%_Contrib']
secondary = total_gdp_all_sectors['Sec_%_Contrib']
teritary = total_gdp_all_sectors['Ter_%_Contrib']        
states = total_gdp_all_sectors['State']

#Plotting all 3 in Stacked Barchart
plt.figure(figsize=(25,10))
p1 = plt.bar(states, primary)
p2 = plt.bar(states, secondary, bottom=primary)
p3 = plt.bar(states, teritary, bottom=np.array(primary)+np.array(secondary))

plt.xlabel('All States Of India')
plt.ylabel('Total GSDP Percentage (%)')

plt.title('Percentage Contribution of Each sector to GSDP')
plt.xticks(states,rotation=50)
plt.yticks(np.arange(0, 110, 10))

plt.legend((p1[0], p2[0], p3[0]), ('primary', 'secondary', 'teritary'))

plt.show()
```
